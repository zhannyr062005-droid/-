```python
import pandas as pd

df = pd.read_excel("insomnia_stress_dataset_N150.xlsx")
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(x='sleep_hours', y='bedtime_hour', data=df)
plt.title('Диаграмма рассеяния: Часы сна против Времени отхода ко сну')
plt.xlabel('Часы сна')
plt.ylabel('Время отхода ко сну')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Cortisol_nmol_L', y='sleep_quality', data=df)
plt.title('Диаграмма рассеяния: Уровень кортизола против Качества сна')
plt.xlabel('Уровень кортизола (нмоль/л)')
plt.ylabel('Качество сна')
plt.grid(True)
plt.show()
```

```python
# 1. Қажетті кітапханалар
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

# 2. Target жасау (инсомния risk)
df["insomnia_risk"] = (
    (df["sleep_hours"] < 6) |
    (df["sleep_latency_min"] > 30) |
    (df["night_awakenings"] >= 2) |
    (df["sleep_quality"] <= 2)
).astype(int)

# 3. X және y бөлу
X = df.drop(columns=["insomnia_risk"])
y = df["insomnia_risk"]

# 4. Масштабтау
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 5. Train/Test бөлу
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# 6. Модель құру
model = LogisticRegression()

# 7. ОҚЫТУ (МАҢЫЗДЫ!)
model.fit(X_train, y_train)

# 8. Accuracy тексеру
accuracy = model.score(X_test, y_test)
print("Accuracy:", accuracy)
```
